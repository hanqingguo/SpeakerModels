```python
# import librosa
import scipy.io.wavfile as wf
import matplotlib.pyplot as plt
import librosa.display
import librosa
```

```python
def drawWav(file):
    wav = wf.read(file)
    sr = wav[0]
    # samples = wav[1]
    samples, sr = librosa.load(file, sr=sr)
    plt.figure()
    plt.subplot(1, 1, 1)
    librosa.display.waveplot(samples, sr=sr)
    plt.title('Monophonic')
```

```python
file1 = "4.wav"
drawWav(file1)
```

```python
file2 = "4-low.wav"
drawWav(file2)
```
